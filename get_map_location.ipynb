```python
import json
import requests

bus_names = ['1', '1A', '2', '2A', '3', '3A', '3AX', '3X', '4', '5', '5AX', '5X', '6A', '6B', '7', '8', '8A', '9', '9A', '10', '10A', '10B', '10X', '11', '12', '15', '16', '16S', '17', '17S', '18', '18A', '18B', '19', '21A', '22', '23', '25', '25AX', '25B', '25BS', '26', '26A', '27', '28A', '28B', '28C', '29', '30', '30X', '32', '33', '34', '35', '35S', '36', '37', '39', '50', '50B', '51', '51A', '51X', '52', '55', '56', '59', '71', '71S', '72', '73', '73S', '101X', '102X', '701X', 'AP1', 'AP1X', 'H1', 'H2', 'H3', 'MT1', 'MT2', 'MT3', 'MT4', 'MT5', 'N1A', 'N1B', 'N2', 'N3', 'N5', 'N6']
color_lists = ['#2DA65D', '#595A38', '#E9DC7B', '#28ECAC', '#C5FC92', '#C641F0', '#942E3C', '#E9DF83', '#674B67', '#C7BACC', '#329432', '#BA32EE', '#674FC9', '#E1D664', '#DA5232', '#D7B1C7', '#642AF7', '#BE9B64', '#A57B5B', '#8A16A5', '#5972B7', '#B8346E', '#E25302', '#5686B4', '#738E38', '#28F3F2', '#46A996', '#E766E2', '#105642', '#2706A9', '#82D219', '#927753', '#5634F8', '#FBEC4A', '#3A0DD4', '#DF900D', '#27E6DF', '#3AC474', '#4B087D', '#8BEC40', '#270FF1', '#8A1D04', '#D20E1B', '#BE3831', '#7E20E0', '#47B991', '#755F56', '#C7E4BA', '#FBD753', '#EB78D5', '#DF72A2', '#EB51CB', '#6A3EA4', '#5545D5', '#43843A', '#E97DB8', '#159FB4', '#9B5486', '#E18101', '#CF84BD', '#E7B8FE', '#FD15D3', '#4CDD84', '#E2E3C4', '#8A0AC6', '#36CBBE', '#357889', '#CED71D', '#A722DB', '#ADF4B4', '#CF6AF0', '#CA754F', '#6971E2', '#748333', '#0730CF', '#85185C', '#51775F', '#893985', '#46F6D3', '#041F12', '#B47DB0', '#8BD10A', '#E62DF9', '#5798C3', '#D35B8C', '#AF7C51', '#29BAE2', '#405249', '#4AB4B3', '#4C11CF', '#BB05D1']

def filterBusLists(stationCode, x):
    if(stationCode == x['stationCode']):
        return True
        
def unique(list1):
    unique_list = []
    for x in list1:
        if x not in unique_list:
            unique_list.append(x)
    return unique_list

def getLocationUrl(bus_names):
    get_location_url_result = []
    for idx,bus_name in enumerate(bus_names):
        location_url = "https://bis.dsat.gov.mo:37812/macauweb/routestation/location"

        for dir_idx in range(0,2):
            
            location_data = {
                "routeName": bus_name,
                "dir": dir_idx
            }
            location_headers = {
            #     "Accept": "application/json, text/javascript, */*; q=0.01",
            #     "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
                "token": "479120230b5118a20118a94e406b10b79999a633ce94"
            }

            location_response = requests.post(location_url, headers=location_headers, data=location_data)
            
            if(location_response.json()['data'] != ""):
                location_results = location_response.json()['data']['stationInfoList']
                for location_result in location_results:
                    location_result_json = {
                        'bus_lists': [bus_name],
                        'stationCode': location_result['stationCode'] ,
                        'coordinate': [ location_result['longitude'], location_result['latitude'] ],
                        'description': location_result['stationName'],
                        'circle_color': color_lists[idx],
                        'circle_radius': 6,
                        'circle_stroke_width': 2,
            #             'circle_stroke_opacity': 0.6,
                        'circle-stroke-color': '#ffffff'
                    } 
                    filter_check = list(filter(lambda x: filterBusLists(location_result['stationCode'], x), get_location_url_result))

                    if(len(filter_check) == 0): #True is mean no this json
                        get_location_url_result.append(location_result_json)
                    else:
                        for enumerate_index, index_key in enumerate(get_location_url_result):
                            if(index_key['stationCode'] == location_result_json['stationCode']):
                                get_location_url_result[enumerate_index]['bus_lists'].append(bus_name)
                                get_location_url_result[enumerate_index]['bus_lists'] = unique(get_location_url_result[enumerate_index]['bus_lists'])
    
    return get_location_url_result

get_location_url_result = getLocationUrl(bus_names)
```
